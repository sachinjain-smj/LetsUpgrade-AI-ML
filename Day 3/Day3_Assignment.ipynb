# Q1. Write a program to subtract two complex numbers in Python

```python
x = complex(input("Enter first complex number:\n"))
y = complex(input("Enter second complex number:\n"))

print("Subtraction of two complex number is",(complex(x-y))) 
```

# Q3. Write a program to swap two numbers in Python with the help of a temporary variable.


```python
x= int(input("Enter a number for x  "))
y= int(input("Enter a number for y  "))

temp=x
x=y 
y=temp

print("Value of x after swap : ", x, " and y : ", y)
```

# Q4. Write a program to swap two numbers in Python without using a temporary variable

```python
x= int(input("Enter a number for x: "))
y= int(input("Enter a number for y: "))

x, y = y, x
print("Value of x after swap : %d \nValue of y after swap : %d"%(x,y))
```

# Q5. Write a program to convert fahrenheit to kelvin and celsius both

```python
f = int(input("Enter the tempaerature in fahrenheit "))

print(f, "Fahrenheit is equal to %.2f Celsius"%((f - 32) * 5/9))

print(f, "Fahrenheit is equal to %.2f Kelvin"%((f - 32) * 5/9 + 273.15))
```
