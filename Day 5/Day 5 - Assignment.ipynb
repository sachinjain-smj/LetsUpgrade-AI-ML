# Question 1 :
Write a Python program to find the first 20 non-even prime natural numbers.

```python
count=0
for i in range(3,100):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i,end=",")
        count+=1
        
    if count==20:
        break
```

# Question 3:
Write a Python program to check if the given string is a Palindrome or Anagram or None of them.
Display the message accordingly to the user.

```python
sample1=input("Enter the string sample 1\n")
sample2=input("Enter the string sample 2\n")

message="no"
test=sample1[::-1]
if sample1.upper()==test.upper():
    print("\n{} is a Palindrome".format(sample1))
    message="yes"
    
if sorted(sample1.lower())== sorted(sample2.lower()):
    print("\n{} is a Anagram of {}".format(sample1,sample2))
    message="yes"
    
if message=="no":
    print("There is no Correlation in strings")
```
