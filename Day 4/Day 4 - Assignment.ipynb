Research on whether addition, subtraction, multiplication, division, floor division and modulus operations be performed on complex numbers. Based on your study, implement a Python program to demonstrate these operations.

# Question 2 :
Research on range() functions and its parameters. Create a markdown cell and write in your own words (no copy-paste from google please) what you understand about it. Implement a small program of your choice on the same


range() function is used to get sequenced list of values in particular defined range.

We can pass 3 in range start, stop and step.

1. Start parameter is passed in range to start the sequence from that number.
2. Stop parameter is given to stop the sequence at n-1. 
3. Step parameter is used for assigning the number of jumps that the sequence should contain in a defined range

Start Parameter and Step parameter are not necessary, we can just pass end parameter as range(interger), and the sequence will start from 0 and take 1 step by default till integer-1. 

For eg.
range(100)
it will return numbers from 0-99
range(5,150,5)
it will return numbers from 5-145 with a step of 5 in between. 


```python
for n in range(5,150,5):
  print(n)
```

# Question 3:
Consider two numbers. Perform their subtraction and if the result of subtraction is greater than
25, print their multiplication result else print their division result.

```python
x=int(input("Enter first number "))
y=int(input("Enter second number "))
z=x-y
print("Subtration of the number is:",z)
if z>25:
    print(x*y)
else:
    print(x/y)
```

# Question 4:
Consider a list of 10 elements of integer values. If the number in the list is divisible by 2, print the
result as "square of that number minus 2".

```python
l=[23,62, 57, 3, 4, 11, 12, 96, 14, 21, 22]
for n in l:
   if n%2==0:
     print((n**2)-2)
```
