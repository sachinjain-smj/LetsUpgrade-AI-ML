```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
hrdataset=pd.read_csv("general_data.csv")
hrdataset.head()
```

```python
hrdataset["Attrition"]=hrdataset.Attrition.map({"Yes":1,"No":0})
```

```python
hrdataset.head()
```

```python
hrdataset.isnull().sum()
```

```python
hrdata=hrdataset.dropna()
```

```python
hrdata.isnull().sum()
```

# 1.Age vs Attrition Hypothesis
Ho = There is no correlation between Age and Attrition
H1 = There is corelation between Age and Attrition

```python
stats,p=pearsonr(hrdata.Attrition,hrdata.Age)
print(stats,p)
```

# Since p<0.05 we reject Ho and accept H1

# 2.DistanceFromHome vs Attrition Hypothesis

# Since p>0.05 we Accept Ho and reject H1
